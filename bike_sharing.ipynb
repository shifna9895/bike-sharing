# ** BIKE SHARING DATASET DATA ANALYSIS**

**A SHORT DESCRIPTION OF THE FEATURES.**

datetime - hourly date + timestamp

season - 1 = spring, 2 = summer, 3 = fall, 4 = winter

holiday - whether the day is considered a holiday

workingday - whether the day is neither a weekend nor holiday

weather -

1: Clear, Few clouds, Partly cloudy, Partly cloudy

2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist

3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds

4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

temp - temperature in Celsius

atemp - "feels like" temperature in Celsius

humidity - relative humidity

windspeed - wind speed

casual - number of non-registered user rentals initiated

registered - number of registered user rentals initiated

count - number of total rentals

```python
# 1) Filtering for Summer Days (season = 1) and Clear Weather (weather = 2)
df[(df['season']==1)&(df['weather']==2)]
```

```python
# 2) Filtering for Working Days with Temperatures between 20C and 30C
df.loc[(df['workingday']==1)&((df['temp']>=20)&(df['temp']<=30))]
```

```python
# 3) Filtering for Holidays with High Humidity (> 50%)
df[(df['holiday']==1)&(df['humidity']>50)]
```

```python
# 4) accessing all rows and columns from datetime to temp using loc
df.loc[:,'datetime':'temp']
```

```python
# 5) accessing first 80 rows and season only
df.loc[0:80,'season']
```

```python
# 6) Filtering for Winter Days with Temperature < 30Â°C and CasualÂ RentalsÂ <Â 30
df.loc[(df['season']==3) & (df['temp']<30) & (df['casual']<30)]
```

```python
# 7) Filtering for Summer Days with Humidity > 50% and WindÂ SpeedÂ <Â 10
df.loc[(df['season']==1) & (df['humidity']>50) & (df['windspeed']<10)]
```

```python
# 8) Filtering for Winter Days (season = 3) with Low Wind Speed (< 20)
df[(df['season'] == 3) & (df['windspeed']<20)]
```

```python
# 9) Filtering for Non-Working Days (workingday = 0) with Casual Rentals > 100
df.loc[(df['workingday'] == 0) & (df['casual']>100)]
```

```python
# 10) Filtering for Registered Rentals between 200 and 300 on Spring Days (season = 1)
df[(df['season'] == 1) & (df['registered'] >= 200) & (df['registered']<=300)]
```

```python
# 11) Filtering for Clear Weather (weather = 1) and Temperature > 25Â°C
df.loc[(df['weather'] == 1) & (df['temp']>25)]
```

```python
# 12) Filtering for Light Rain or Snow (weather = 3) on Holidays
df.loc[(df['weather'] == 3) & (df['holiday']==1)]
```

```python
# 13) Filtering for Working Days with Rentals Count > 500 and Humidity < 70%
df.loc[(df['workingday'] == 1) & (df['count']>500) &(df['humidity']<70)]
```

```python
# 14) Filtering for Summer Days with Light Wind (windspeed < 5) and High Rentals (> 700)
df.loc[(df['season'] == 2) & (df['windspeed'] < 5) & (df['count']>700)]
```

```python
# 15) Filtering for Winter Days with Casual Rentals < 50 and Humidity > 85%
df.loc[(df['season'] == 4) & (df['casual'] < 50) & (df['humidity'] > 85)]
```

```python
# 16) Filtering for Working Days with Temperatures < 10Â°C and Registered Rentals > 300
df.loc[(df['workingday'] == 1) & (df['temp'] < 10) & (df['registered']>300)]
```

```python
# 17) Filtering for Fall Days (season = 3) with Light Rain or Snow (weather = 3)
df.loc[(df['season'] == 3) & (df['weather']==3)]
```

```python
# 18) Filtering for Non-Working Days with Registered Rentals between 100 and 200
df.loc[(df['workingday'] == 0) & (df['registered'] >= 100) & (df['registered']<=200)]
```

```python
# 19) Filtering for Misty or Cloudy Weather (weather = 2) with Wind Speed > 15
df.loc[(df['weather'] == 2) & (df['windspeed']>15)]
```

```python
# 20) Filtering for Spring Days (season = 1) with Registered Rentals > 400 and Humidity < 60%
df.loc[(df['season'] == 1) & (df['registered'] > 400) & (df['humidity']<60)]
```
